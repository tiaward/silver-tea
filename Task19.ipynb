```python
# Importing various libraries including sklearn
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression 
```

```python
# Importing the data from the csv file
df = pd.read_csv('insurance.csv')
```

```python
# Using the plt and sns libraries to create a scatter graph with age as the x axis and charges as the y
plt.figure()
sns.scatterplot(data=df,x='age',y='charges')
plt.show()
plt.close()
```

```python
linreg_model = LinearRegression()
```

```python
# Reshaping the inputs
X = df['age'].values.reshape(-1,1)
y = df['charges'].values
```

```python
# Fitting the data with prediction
linreg_model.fit(X,y)
y_pred= linreg_model.predict(X)
```

```python
# Again printing the re-done graph with line of best fit 
plt.figure()
sns.scatterplot(data=df,x='age',y='charges')
plt.plot(X, y_pred)
plt.show()
plt.close()
```

This is the finished model with line of best fit. The prediction would be that the line of best fit 
is the average charges of each age group, meaning that as the person ages, the average charges will increase at a slow rate seeing as the the line slope is fairly low and flat.
